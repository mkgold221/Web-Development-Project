```python
df['Total Sales'] = df['Total Sales'].replace('In-Store', 0)
```

```python
df['Price per Unit'] = df['Price per Unit'].astype(str).str.replace(r'\$', '', regex=True)
```

```python
df['Total Sales'] = df['Total Sales'].astype(str).str.replace(r'\$', '', regex=True)
```

```python
df['Sales Method'].replace('01/19/2022', 'Online', inplace=True)
```

```python
df['Invoice Number'].replace(0,'INV-1005',inplace=True)
```

```python
df['Date'].mode()
```

```python
# Clean and convert Price per Unit
df['Price per Unit'] = (
    df['Price per Unit']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)
    .str.strip()
    .replace('', '0')
    .astype(float)
)

# Clean and convert Units Sold
df['Units Sold'] = (
    df['Units Sold']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)
    .str.strip()
    .replace('', '0')
    .astype(float)
)

# Clean and convert Total Sales
df['Total Sales'] = (
    df['Total Sales']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)
    .str.strip()
    .replace('', '0')
    .astype(float)
)
```

```python
df.head()
```

```python
df['Total Sales'] = df['Price per Unit'] * df['Units Sold']
```

```python
df
```

```python
import pandas as pd

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # ensure datetime dtype
df['Date'] = df['Date'].fillna(pd.to_datetime('03/12/2022'))
```

```python
# First, install the pyspellchecker package
!pip install pyspellchecker

# Then import and use it
from spellchecker import SpellChecker

# Create spell checker
spell = SpellChecker()

# Pick the column to check
column = 'City'

# Find words that might be misspelled
misspelled = set()

for text in df[column].dropna():
    for word in str(text).split():
        if word.lower() not in spell:  # Optional: lowercase check
            misspelled.add(word)

print("Possible misspellings:", misspelled)
```
